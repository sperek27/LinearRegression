```python
import numpy as np
import matplotlib.pyplot as plt

#Generate a linearly correlated dataset with size n  
n = 100
#Create the coefficients of a linear equation
Coeffs = np.random.normal(size=2) + np.random.randint(low=-5,high=5,size=2)
#Randomly select n x-values
X = np.random.normal(size=n) + np.random.randint(low=1,high=5,size=n)
#Calculate the y-values for each x coordinate then add random noise
Y = [Coeffs[0]*X[n]+Coeffs[1] for n in range(n)] + np.random.normal(size=n) + np.random.randint(low=-3,high=3,size=n)

plt.scatter(X,Y)
plt.plot([min(X), max(X)], [min(X)*Coeffs[0] + Coeffs[1], max(X)*Coeffs[0] + Coeffs[1]], color='purple')
plt.show()
```

```python
#Learned coeffecients and learning rate
m = 0
b = 0
L = 0.001
loss = [] 

for i in range(3000):
    Y_pred = m*X + b
    #Calculate the partial derivatives of the loss function with respect to m and b
    D_m = (-2/n) * sum(X * (Y - Y_pred))
    D_b = (-2/n) * sum(Y - Y_pred)
    #Update m and b in the direction of loss minimums
    m -= L * D_m
    b -= L * D_b
    #Record loss
    if i<250:
        loss.append((sum((Y - Y_pred)**2)/n))
#Show accuracy vs iterations
plt.plot(loss)
plt.show()
```

```python
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(X)*m + b, max(X)*m + b], color='red')
plt.show()
```
